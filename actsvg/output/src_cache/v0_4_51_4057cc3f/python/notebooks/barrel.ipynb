# Example: plotting a barrel detector 

This examples shows how to plot a cylindrical barrel detector. It loads pologon definitions from the common data directory and then plots the created surfaces in various different options.

First, however, import the actsvg modules:

```python
# This will be needed to display the SVGs in the notebook
from IPython.core.display import SVG

# This is the actsvg module
import actsvg
from actsvg import core, examples, proto, display, style
```

Load the barrel modules 

```python
module_fill = style.defaults.sensitive_fill()
module_stroke = style.default.sensitive_stroke()
# This reads from 'data/odd_pixel_barrel.hpp' and returns a list of polygons
barrel_polygons = examples.generate_barrel_modules()
# This part converts the point clouds into surface objects
barrel_modules = []
for i, polygon in enumerate(barrel_polygons):
    barrel_modules.append(proto.surface.polygon_from_vertices('module_'+str(i), polygon, module_fill, module_stroke))
```

Select a range, this is a e.g. a slice in z

```python
modules_xy = display.surfaces(barrel_modules, 'xy')
```

Create a temporary file and display, load it back in and display it.

```python
m_xy_file = actsvg.io.file()
m_xy_file.add_objects(modules_xy)
m_xy_file.write("barrel_modules_xy.svg")
```
